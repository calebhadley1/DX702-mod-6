## Use Euclidean distance to find the ATE for this table by selecting matches.

```python
import numpy as np

#Row  D  Z  Y
#1    0  1  10
#2    0  2  14
#3    1  1  18
#4    1  2  22
#5    1  2  25

# Propensity scores:
# Row 1: 0.5
# Row 2: 0.667
# Row 3: 0.5
# Row 4: 0.667
# Row 5. 0.667

h = 1
long_dist = (0.5 - 0.667) / h
assert np.abs(long_dist) < 1

w = (1 / (1 + (1 - long_dist**2))) # 3/4 term cancels in weight
v3 = 18 - 10 * w - 14 * (1 - w)
v4 = 22 - 14 * w - 10 * (1 - w)
v5 = 25 - 14 * w - 10 * (1 - w)
effect_att = (v3 + v4 + v5) / 3

wa = (1 - 0**2) / ((1 - 0**2) + (1 - long_dist**2) + (1 - long_dist**2))
wb = (1 - long_dist**2) / ((1 - 0**2) + 2 * (1 - long_dist**2))
effect_1 = wa * 18 + wb * 22 + wb * 25 - 10
wc = (1 - long_dist**2) / ((1 - 0**2) + (1 - 0**2) + (1 - long_dist**2))
wd = (1 - 0**2) / ((1 - 0**2) + (1 - 0**2) + (1 - long_dist**2))
effect_2 = wc * 18 + wd * 22 + wd * 25 - 14

effect_atut = (effect_1 + effect_2) / 2

print(effect_att, effect_atut, effect_atut * 0.4 + effect_att * 0.6)
```
